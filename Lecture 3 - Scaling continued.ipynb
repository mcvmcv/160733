## Chemical Reactor problem
Consider now the problem of modelling a chemical reaction that occurs within a continously stirred tank. Continuously stirred, or well-mixed, means that the concentration is kept constant throughout the tank. The chemical comes into the tank at concentration $c_i$, specified as mass per volume, and leaves at concentration $c(t)$. The flow in and out of the tank are assumed to be equal. The tank is depicted below:


```python
Image("chemicalreaction.png")
```

We assume that the chemical reaction occurs according to the law of mass action, *i.e.* $r = kc$ where the reaction rate $r$ has dimensions of mass per unit volume per unit time, and $k$ of inverse time.

Then, by conservation of mass,
\begin{align}
    \frac{d}{dt}(Vc(t)) &= qc_i - qc(t) - Vkc(t) \nonumber \\
    \frac{dc}{dt} &= \frac{q}{V}(c_i - c) - kc, \quad t > 0 \\
    c(0) &= c_0.
\end{align}

To nondimensionalise this, we need to choose dimensionless independent and dependent variables. So we need a characteristic concentration and time, usually formed from the constants in the problem, $c_i, c_0, V, q, k$. In our case, either $c_i$ or $c_0$ will do for the concentration, and to form a time scale we can either choose characteristic time $T = \frac{1}{k}$ or $T = \frac{V}{q}$. We won't make this choice yet, and just leave it as $T$. So, 
\begin{align*}
    C &= \frac{c}{c_i} & \tau = \frac{t}{T}.
\end{align*}
Noting that
\begin{equation*}
    \frac{dc}{dt} = \frac{c_i}{T}\frac{dC}{d\tau},
\end{equation*}
We now have the dimensionless model
\begin{align*}
    \frac{dC}{d\tau} &= \frac{qT}{V}(1 - C) - kTC, \quad \tau > 0,\\
    C(0) &= \gamma,
\end{align*}
where $\gamma = \frac{c_0}{c_i}$. Notice that if we choose the timescale $T = V/q$, then our model becomes
\begin{align}
    \frac{dC}{d\tau} &= 1 - C - \beta C, \quad \tau > 0,  \\
    C(0) &= \gamma \nonumber
\end{align}
where $\displaystyle \beta = \frac{kV}{q}$, whereas if we choose $T = 1/k$, we get
\begin{align}
    \frac{dC}{d\tau} &= \frac{1}{\beta}(1 - C) - C, \quad \tau > 0, \\
    C(0) &= \gamma \nonumber
  \end{align}

So, why would we choose one of these systems over the other? Essentially, it depends on the relative magnitude of the terms. For example, if the reaction was slow compared with the the flow through the reactor. This would somehow mean that $k$ is small compared with the flow rate $q/V$. So we would write $k \ll q/V$, or alternatively $V/q \ll 1/k$. So our dimensionless parameter $\beta$ is small, or $\beta \ll 1$. So in this case, we should scale time according to the flow rate, i.e. use the first scaling. Things that happen at nondimensional times of order 1 are happening in real world times of order $V/q$, i.e. on a very short timescale.

Conversely, if the reaction is fast compared with the flow rate, then $\beta \gg 1$, and we should scale time with the reaction time scale, putting the small parameter $1/\beta$ correctly on the flow rate term.

Sometimes we might want to approximate our problem by deleting small terms in the equation. This is a pretty common thing to want to do, as it can give equations that are analytically solvable from nonlinear ones. For example, if $\beta \ll 1$, we can drop the small term off our first rescaling, giving
\begin{align*}
    \frac{dC}{d\tau} &= 1 - C, \quad C(0) = \gamma
\end{align*}
This is analytically solvable to give $C(\tau) = 1 + (\gamma - 1)e^{-\tau}$. This is believable, the concentration over time approaches the concentration of the feed.

On the other hand, if we used the longer timescale ($T = 1/k$) scaling we would have
\begin{align*}
    \beta \frac{dC}{d\tau} = (1 - C) - \beta C
\end{align*}
If we used this nondimensionalisation by accident, then dropping off the $\beta$ terms would give us the approximation $C = 1$, which does not satisfy the initial conditions, but yet does in some sense satisfy the long term behaviour of the differential equations. So we need to be careful! We'll learn how to deal with these situations later in the course.

# The projectile problem
This lecture we're going to focus on one particular example, the
*Projectile problem*. The projectile is thrust vertically upward from the
earth. The main assumption that we're going to make is that air resistance
is negligible. This is a normal kind of *first approximation* - if it
then turns out that a model doesn't match experimental data, we start
adding in more odds and ends.

By Newton's second law, force = mass times acceleration:
\begin{equation*}
  m \frac{d^2 h}{dt^2} = -G \frac{Mm}{(h+R)^2}
\end{equation*}
where $G$ is the universal gravitation constant, $R$ the radius of the planet, $M$ the mass of the planet, $m$ the mass of the projectile, and $h$ the height above the earth. Note that when $h = 0$,
gravitational force = $-mg$, where 
\begin{equation*}
  g = \frac{GM}{R^2}
\end{equation*}
So we'll rewrite the equation as
\begin{equation}
  \frac{d^2h}{dt^2} = -\frac{R^2 g}{(h + R)^2}
\end{equation}
and pose initial conditions
\begin{equation}
  h(0) = 0, \quad \frac{dh}{dt}(0) = V.
\end{equation}
We'll take these two equations as our mathematical model of the problem. So, let's now embark on dimensional
analysis:

##Dimensional Analysis
\begin{align}
  [t] &= [T], & [V] &= [L][T]^{-1} \\
  [h] &= [L], & [g] &= [L][T]^{-2} \\
  [R] &= [L], 
\end{align}
Note we only have $L$ and $T$ again as fundamental dimensions. So, if $\pi$
is a dimensionless combination of $t, h, R, V, g$, we have
\begin{equation*}
  [\pi] = [t^{\alpha_1}h^{\alpha_2}R^{\alpha_3}V^{\alpha_4}g^{\alpha_5}] =
  [T]^{\alpha_1 - \alpha_4 - 2\alpha_5}[L]^{\alpha_2 + \alpha_3 + \alpha_4 +
  \alpha_5} = 1.
\end{equation*}
\emph{i.e.}
\begin{align*}
  \alpha_1 - \alpha_4 - 2\alpha_5 &= 0, \\
  \alpha_2 + \alpha_3 + \alpha_4 + \alpha_5 &= 0
\end{align*}
This system has rank 2, so there are three independent dimensionless
quantities (and lots of different ways of generating them). It's good if
our variables $t, h$ appear only once, so a good selection is
\begin{align*}
  \pi_1 &= \frac{h}{R}, & \pi_2 &= \frac{t}{R/V}, & \pi_3 &=
  \frac{V}{\sqrt{gR}}
\end{align*}
By the Pi theorem, if there is a physical law relating $t, h, R, V, g$,
there must be an equivalent one relating $\pi_1, \pi_2, \pi_3$. In
particular, we could write it as
\begin{equation*}
  \frac{h}{R} = f\left(\frac{t}{R/V}, \frac{V}{\sqrt{gR}}\right)
\end{equation*}
for some function $f(\pi_2, \pi_3)$. We can get some useful stuff out of
this. Notice that $t$ only appears in the first argument. So it we wanted
to find the maximum height for a given initial velocity, we could
differentiate and set $h'(t) = 0$ to give
\begin{align*}
  \frac{\partial f}{\partial \pi_2}\left(\frac{t_\text{max}}{R/V},
  \frac{V}{\sqrt{gT}}\right) = 0 \\
  \text{or,}\qquad
  \frac{t_\text{max}}{R/V} = F\left(\frac{V}{\sqrt{gR}}\right)
\end{align*}
In other words, the maximum height depends only on the dimensionless
combination $\frac{V}{\sqrt{gR}}$. By fitting $F$ from data from one planet
(with a particular $g$ and $R$, we can automatically know the relationship
for all planets (as long as we know their $g$ and $R$)).

## Scaling
Now for the tricky bit: scaling the equations. We need characteristic $t_c$
and $h_c$ so we can set
\begin{align*}
  \bar{t} &= \frac{t}{t_c} & \bar{h} &= \frac{h}{h_c}.
\end{align*}
Length scale options: $R$, $V^2/g$. Timescale options: $R/V$, $\sqrt{R/g}$,
and $V/g$. hmmm. Here are three possibilities:
\begin{align}
  \bar{t} &= \frac{t}{R/V}, & \bar{h} &= \frac{h}{R} \\
  \bar{t} &= \frac{t}{\sqrt{R/g}}, & \bar{h} &= \frac{h}{R} \\
  \bar{t} &= \frac{t}{Vg^{-1}}, & \bar{h} &= \frac{h}{V^2g^{-1}}
\end{align}
which lead to dimensionless problems, equivalent to our original one:
\begin{align}
  \epsilon \frac{d^2 \bar{h}}{d\bar{t}^2} &= -\frac{1}{(1 + \bar{h})^2}, &
  \bar{h}(0) &= 0, & \frac{d\bar{h}}{d\bar{t}}(0) &= 1, \\
  \frac{d^2 \bar{h}}{d\bar{t}^2} &= -\frac{1}{(1 + \bar{h})^2}, &
  \bar{h}(0) &= 0, & \frac{d\bar{h}}{d\bar{t}}(0) &= \sqrt{\epsilon}, \\
  \frac{d^2 \bar{h}}{d\bar{t}^2} &= -\frac{1}{(1 +
    \epsilon\bar{h})^2}, &
  \bar{h}(0) &= 0, & \frac{d\bar{h}}{d\bar{t}}(0) &= 1, \\
\end{align}
where
\begin{equation*}
  \epsilon = \frac{V^2}{gR}.
\end{equation*}
Now what happens if $\epsilon$ is small (small $V$). Then, one may be
tempted to remove terms from the approximation.

THe first one has no solution, the second has no phyiscal solution (concave
down, $\bar{h}(0) = 0$, and the third is OK, gives
\begin{equation*}
  \bar{h} = \bar{t} - \bar{t}^2 / 2
\end{equation*}
or $h = -\frac{1}{2}gt^2 + Vt$.

If $V$ small, $g$ constant, decelerates to zero in time $V/g$. distance is
$V/g$ times average velocity $V/2$, so $V^2/(2g)$ is good distance scale.


```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# problem parameters
R = 6371e3 # radius of the earth in m
V = 500 * 1000 / 3600 # initial velocity (km/h converted to m/s)
M = 5.972e24
G = 6.67408e-11
g = G*M/(R**2) # m/s^2
epsilon = V**2/g/R
h0 = 0
v0 = V

print "epsilon = {:2f}".format(epsilon)

# We'll solve the full problem, and see how the terms in the different scalings compare
def rhs(y, t):
    h = y[0]
    v = y[1]
    
    dh = v
    dv = -g * (R / (h + R))**2
    return [dh, dv]

t = np.linspace(0, 2.1*V/g, 1000)
dt = np.mean(np.diff(t))
soln = odeint(rhs, [h0, v0], t)
h = soln[:, 0]
v = soln[:, 1]
a = np.gradient(v, dt)
```

### First scaling

```python
tc = R/V
hc = R

hbar = h / hc
tbar = t / tc

lhs = epsilon * tc**2 / hc * a
rhs = -1 / (1 + hbar)**2
plt.plot(tbar, lhs, tbar, rhs)
plt.legend(('lhs', 'rhs'))
max(lhs)
```

Note that the right and left hand sides are both in exactly the same scale (both near $-1$), so it makes no sense to delete the left hand side of the equation. Containing an epsilon factor is a red herring, as the second derivative term is large. You can also see that the time scale is inappropriate, most of the action happens before $\bar{t}$ gets to $0.0007$.

### Second scaling
This one is a little different. Here we are not dropping terms from an equation, but removing the physics from the problem -- saying that it's initial speed is 0 rather than a positive number. This immediately should raise a red flag

### Third scaling
This time the term we are considering deleting is $\epsilon\bar{h}$, and the term we should compare it to is $1$. 

```python
tc = V/g
hc = V**2/g

hbar = h / hc
tbar = t / tc

plt.figure()
plt.plot(tbar, hbar)
plt.ylabel('${\\bar h}$')
plt.xlabel('$\\bar{t}$')

plt.figure()
ax = plt.gca()
plt.plot(tbar, epsilon*hbar, tbar, np.ones(len(tbar)))
ax.set_ylim((-0.1, 1.1))
ax.legend(('$\epsilon\\bar{h}$', 1))
plt.xlabel('$\\bar{t}$')
```

You can see that $\bar{h}$ is nicely in scale, as is $\bar{t}$, and the term we were assuming was small, definitely is. So dropping it shouldn't be a problem. Finally, for completeness, we solve our modified equation and see how well it compares.

```python
hbar = tbar - tbar**2/2

plt.plot(t, h, tc*tbar, hc*hbar)
plt.ylabel('h')
plt.xlabel('t')
plt.legend(('original solution', 'approximate solution'), loc='best')
```
